Suppose  a  random  sample  of  10  manufactured  items  is  selected.  Answer  the  following questions. 

### 2.A. Probability that none of the items are defective?


### Importing all the Libraries 

```python
# Libraries for Data Handling and Manipulation
import numpy as np 
import pandas as pd

# Libraries for Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline

# Numeric Operations and Mathematical Functions
import math

# Statistical Analysis
import scipy.stats as stats
from scipy.stats import ttest_ind

# Set display option to restrict float values to 3 decimal places
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# declare the sample size in variable 'n' which represents the number of manufactured itsem selected randomly
n=10
```

```python
# declare 'p' which represents the probability of success, i.e.,the failure rate for the manufactured item 
p=0.05
```

```python
# declare different possible number of manufactured items selected which non of them are defective in a numpy array
k = np.arange(0,11)
k
```

P(X=x)=(n/x)px(1p)nx

```python
# import the required function
from scipy.stats import binom

# use the binom.pmf() function to generate the probability distribution
binomial = binom.pmf(k=k, n=n, p=p)
```

```python
binomial
```

```python
# plot the distribution
plt.bar(k, binomial) # make a bar plot
plt.title("Binomial: n=%i , p=%.2f" % (n, p), fontsize=15) # set the title
plt.xlabel("Number items manufactured") # set the x-axis label
plt.ylabel("Non of the item being defective") # set the y-axis label
plt.show() # display the plot
```

```python
binomial[0]
```

### B. Write the null hypothesis and alternative hypothesis.

The null hypothesis for the two-sample t-test is that there is no significant difference between the means of the two groups. In this case, it would be:

H0: 1 = 2

where 1 is the mean funds raised by companies that are still operating and 2 is the mean funds raised by companies that closed down.

The alternative hypothesis is that there is a significant difference between the means of the two groups. In this case, it would be:

Ha: 1  2

where 1 is the mean funds raised by companies that are still operating and 2 is the mean funds raised by companies that closed down.


```python
# Extract data for the t-test
operating = df3.loc[df1['OperatingState'] == 'Operating', 'Funds_in_million']
closed = df3.loc[df1['OperatingState'] == 'Closed', 'Funds_in_million']

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(Operating, Closed, equal_var=False)

# Write the null and alternative hypotheses
null_hypothesis = "There is no significant difference between the mean funds raised by companies that are still operating and companies that closed down."
alternative_hypothesis = "There is a significant difference between the mean funds raised by companies that are still operating and companies that closed down."

# Print the null and alternative hypotheses
print("Null Hypothesis: " + null_hypothesis)
print("Alternative Hypothesis: " + alternative_hypothesis)
```

### C. Test for significance and conclusion

```python
# Conduct a two-sample t-test
t_stat, p_val = stats.ttest_ind(operating, closed)

# Print the t-statistic and p-value
print("t-statistic: {:.3f}".format(t_stat))
print("p-value: {:.3f}".format(p_val))

# Draw a conclusion based on the results
if p_val < 0.05:
    print("We reject the null hypothesis.")
    print("There is a significant difference between the mean funds raised by companies that are still operating and companies that closed down.")
else:
    print("We fail to reject the null hypothesis.")
    print("There is no significant difference between the mean funds raised by companies that are still operating and companies that closed down.")
```
